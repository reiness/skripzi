```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = [
    "Bicycle", "Boat", "Bottle", "Bus", "Car", "Cat",
    "Chair", "Cup", "Dog", "Motorbike", "People", "Table"
]
image_counts = [652, 679, 547, 527, 638, 735, 648, 519, 801, 503, 609, 505]

# Generate a list of colors for each bar
colors = plt.cm.tab20(np.linspace(0, 1, len(categories)))

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, image_counts, color=colors)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, str(height),
             ha='center', va='bottom', fontsize=9)

plt.xlabel('Object Classes')
plt.ylabel('Number of Images')
plt.title('Number of Images per Class')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your data
categories = [
    "Bicycle", "Boat", "Bottle", "Bus", "Car", "Cat",
    "Chair", "Cup", "Dog", "Motorbike", "People", "Table"
]
image_counts = [21, 12, 26, 3, 14, 23, 119, 114, 12, 18, 103, 68]
frames       = [ 6,  7,  3,  5,  6,  6,   5,   6,  3,   6,   6,    3]

x = np.arange(len(categories))
width = 0.8    # full width for base bars
overlay = 0.4  # narrower overlay for frames

# Colors
base_color    = "#4682B4"   # steelblue
blue_overlay  = "#B0C4DE"   # lightsteelblue
red_overlay   = "#F08080"   # lightcoral, for Bicycle & Motorbike

fig, ax = plt.subplots(figsize=(12, 6))

for i, (cat, img_ct, frm_ct) in enumerate(zip(categories, image_counts, frames)):
    # 1) Base bar
    bar_img = ax.bar(
        x[i], img_ct,
        width=width,
        color=base_color,
        label="Instances" if i == 0 else ""
    )[0]
    # annotate instance count at top-left
    img_x = bar_img.get_x() + 0.02
    img_y = img_ct + 1
    ax.text(
        img_x, img_y,
        str(img_ct),
        ha="left", va="bottom",
        fontsize=10,
        fontweight="semibold"
    )

    # 2) Decide overlay color based on category
    this_overlay = red_overlay if cat in ("Bicycle", "Motorbike") else blue_overlay

    # 3) Overlay bar (frames)
    bar_frm = ax.bar(
        x[i], frm_ct,
        width=overlay,
        color=this_overlay,
        hatch="//",
        label="Frames" if i == 0 else ""
    )[0]
    # annotate frame count at top-right
    frm_x = bar_frm.get_x() + bar_frm.get_width() - 0.02
    frm_y = frm_ct + 1
    ax.text(
        frm_x, frm_y,
        str(frm_ct),
        ha="right", va="bottom",
        fontsize=10,
        fontweight="light"
    )

# Final touches
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha="right")
ax.set_ylabel("Count")
ax.set_title("Image Instances vs. Frames per Class")
ax.legend(loc="upper right")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your data
categories = [
    "Bicycle", "Boat", "Bottle", "Bus", "Car", "Cat",
    "Chair", "Cup", "Dog", "Motorbike", "People", "Table"
]
image_counts = [21, 12, 26, 3, 14, 23, 119, 114, 12, 18, 103, 68]
frames       = [ 6,  7,  3,  5,  6,  6,   5,   6,  3,   6,   6,    3]

x = np.arange(len(categories))
width = 0.8   # both bars will use this full width

# Colors
base_color   = "#4682B4"   # steelblue
blue_overlay = "#B0C4DE"   # lightsteelblue
red_overlay  = "#F08080"   # lightcoral, for Bicycle & Motorbike

fig, ax = plt.subplots(figsize=(12, 6))

for i, (cat, img_ct, frm_ct) in enumerate(zip(categories, image_counts, frames)):
    # 1) Draw the base bar for instances
    bar_img = ax.bar(
        x[i], img_ct,
        width=width,
        color=base_color,
        label="Instances" if i == 0 else ""
    )[0]
    # annotate instance count at top-left
    img_x = bar_img.get_x() + 0.02
    img_y = img_ct + 1
    ax.text(img_x, img_y,
            str(img_ct),
            ha="left", va="bottom",
            fontsize=8)

    # 2) Choose overlay color
    overlay_color = red_overlay if cat in ("Bicycle", "Motorbike") else blue_overlay

    # 3) Draw the overlay bar for frames (same width, semi-opaque)
    bar_frm = ax.bar(
        x[i], frm_ct,
        width=width,
        color=overlay_color,
        alpha=0.6,        # makes underlying bar visible
        hatch="//",
        label="Frames" if i == 0 else ""
    )[0]
    # annotate frame count at top-right
    frm_x = bar_frm.get_x() + bar_frm.get_width() - 0.02
    frm_y = frm_ct + 1
    ax.text(frm_x, frm_y,
            str(frm_ct),
            ha="right", va="bottom",
            fontsize=10,
            fontweight="bold")

# Final styling
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha="right")
ax.set_ylabel("Count")
ax.set_title("Image Instances vs. Frames per Class")
ax.legend(loc="upper right")

plt.tight_layout()
plt.show()
```
